```python
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/sriku2412/GneAI---Digital-Earth-Gardians/main/AI_EarthHack_Dataset_2.csv')
data.head()
```

```python
rank_counts = data['rank'].value_counts().reset_index()
rank_counts.columns = ['Rank', 'Count']

# To display the table
print(rank_counts)
```

```python
# Assuming your dataset is named 'data'
data = data[data['rank'] <= 6]
data = data[data['rank'] > 0]
```

```python
data[data['rank']==6].sample(3)
```

```python
data['category'].value_counts().reset_index()
```

```python
import openai

# Assuming you have a DataFrame 'data' with a 'solution' column
# Define a function to generate summaries using the OpenAI API
def generate_summary(text):
    response = openai.Completion.create(
        engine="gpt-3.5-turbo-instruct",  # Use "davinci" as the engine for text tasks
        prompt=text,
        max_tokens=50,  # Adjust max_tokens as needed for your summary length
    )
    summary = response.choices[0].text
    return summary

# Apply the function to create the 'summary_solution' column
data['summary_solution'] = data['solution'].apply(lambda x: generate_summary(x))

data.head()
```

```python
# Assuming you have a DataFrame 'data' with a 'solution' column and want to create a 'summary_solution' column
data['summary_problem'] = data['problem'].apply(lambda x: generate_summary(x))
data.head()
```
